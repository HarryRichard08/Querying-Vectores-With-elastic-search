```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import pandas as pd
import re
import spacy
#Import all the dependencies
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
from elasticsearch import Elasticsearch
import nltk
stop_words = nltk.corpus.stopwords.words('english')
```

```python
es = Elasticsearch([{'host': 'localhost', 'port': 9200}])
```

```python
model= Doc2Vec.load("d2v.model_updated_one")
```

```python
def query_index():
    new_question = input()
    pattern = re.compile(r'[\W|_]+')
    filtered_string = re.sub(pattern, " ", new_question)
    words = filtered_string.lower().split(' ')
    for word in stop_words:
        while word in words:
            words.remove(word)
    sentence = ' '.join(words)
    
    test_data = word_tokenize(sentence.lower())
    query_vector  =  model.infer_vector(test_data)
#    query_vector = query_vector[0].tolist()
    #print("LEN-",len(query_vector))
    num_outputs=5
    query = {
    "script_score": {
        "query": {"match_all": {}},
        "script": {
        "source": "cosineSimilarity(params.query_vector, 'ticket_vector') +1.0",
        "params": {"query_vector": query_vector}
        }
    }
    }

    body = {
        "size": num_outputs,
        "query": query,
    }

    output = es.search(body=body,index="new_index_one")
#     print(output)
#     print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
#     if len(output['hits']['hits']) >= 4:
#         iter_len = 5
#     else:
#         iter_len = len(output['hits']['hits'])
    for i in output['hits']['hits'] :
        
#         tmp = output['hits']['hits'][i]
        print("#"*100)
#         print("I-",i)
        #print("Top -",(i+1),": Score -",output['hits']['hits'][i]['_score'])
        #print("question-",(tmp['_source']['question']))

        question = i['_source']['ticket']
        issue_id = i['_source']['issue_id']
        issue_key = i['_source']['issue_key']
        score = i["_score"]
        print(question)
        print(issue_id)
        print(issue_key)
        print(score)
```
